```python
import networkx as nx
```

```python
import csv
```

```python
graph = nx.read_gexf("gard_hp.gexf")
```

```python
print(nx.info(graph))
```

```python
w = nx.pagerank_numpy(graph)
```

```python
with open('page_rank_weights_small.csv', 'w') as f:
    for key in w.keys():
        f.write("%s,%s\n"%(key,w[key]))
```
